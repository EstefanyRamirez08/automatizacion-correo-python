## Buscar los datos de las acciones

```python
import yfinance
```

```python
data = yfinance.Ticker("AAPL").history("6mo")
cierre = data.Close
cierre.plot()
```

## Hacer el análisis solicitado por el gestror

```python
maxi = round(cierre.max(),2)
```

```python
minima = round(cierre.min(),2)
```

```python
valor_medio = round(cierre.mean(),2)
```

```python
print(maxi)
print(minima)
print(valor_medio)
```

## Enviar correo electrónico automáticamente
* Abre una nueva pestaña
* Entra al correo
* Haz click en el boton redactar
* Escribimos la dirección del destinatario
* Escribimos el asunto
* Redactamos el correo
* Haz click en "enviar"

```python
from dotenv import load_dotenv
import os
import pyautogui
import pyperclip
import webbrowser
import time
```

```python
load_dotenv()

URL = os.getenv("EMAIL_URL")
webbrowser.open(URL)

destinatario = "micorreo@gmail.com"
Asunto = "Análisis de Acciones"
cuerpo = f"""
Hola 
Buenas Tardes XXX,

Acá te envío el análisis de las acciones de los ultimos 6 meses de Apple.


Cotización máxima: USD {maxi}
Cotización mínima: USD {minima}
Valor medio: USD {valor_medio}

¡Estoy pendiente a cualquier observación!

Saludos,

xxxxxxxxxx

"""

time.sleep(3)
pyautogui.PAUSE = 3
#utilizar el hadware
#dar click en redactar
pyautogui.click(121, 257)

#Guarda informacion en el portapales del compu

pyperclip.copy(destinatario)
pyautogui.hotkey("ctrl","v")

pyautogui.hotkey("tab")
pyperclip.copy(Asunto)
pyautogui.hotkey("ctrl","v")

pyautogui.hotkey("tab")

pyperclip.copy(cuerpo)
pyautogui.hotkey("ctrl","v")

pyautogui.click(378,979)

pyautogui.hotkey("ctrl","w")






```
